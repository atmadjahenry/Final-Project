```python
## Terdapat data yang hanya berisikan missing value, data tersebut dapat dihilangkan
test = test[test.isna().all(axis=1)==False]
```

```python
test.columns = test.columns.str.lower()
```

```python
test[test['credit score']>850].head()
```

```python
## Handle credit score > 850
test['credit score'] = np.where(test['credit score']>850, test['credit score']/10, test['credit score'])
```

```python
# binning credit score
def binning_credscore(value):
    if 850.0 >= value >= 720.0 :
        result = 'Excellent'
    elif value >= 690.0 :
        result = 'Good'
    elif value >= 630.0 :
        result = 'Fair'
    elif 629.0 >= value >= 300.0  :
        result = 'Poor'

    else :
        result = 'credit score out of range'
    return result
```

```python
test['credit score'] = test['credit score'].dropna().apply(binning_credscore)
```

```python
# Handle current loan amount yg nilainya 99999999
test['current loan amount'].replace(99999999,np.nan, inplace=True)
```

```python
# mengubah tipe data years in current job dari string menjadi float
test['years in current job'] = test['years in current job'].str.extract(r"(\d+)")
test['years in current job'] = test['years in current job'].astype(float)
```

```python
test['purpose'].value_counts()
```

```python
# Mengabungkan data other dan Other di kolom purpose
test['purpose'] = test['purpose'].replace('Other','other')
test['purpose'] = test['purpose'].replace('Take a Trip', 'vacation')
```

```python
# Mengabungkan data HomeMortgage dan Have Mortgage di kolom home ownership
test['home ownership'] = test['home ownership'].replace('HaveMortgage','Home Mortgage')
```

## Handling Missing Value

```python
test.isna().sum()
```

```python
# Kolom months since last delinquent memiliki missing value diatas 50% dari data, oleh sebab itu kolom tersebut dapat dihapus

test.drop('months since last delinquent', axis=1, inplace=True)
```

```python
# Kolom Tax Liens dan Bankruptcies memiliki sedikit missing value, drop missing value yang berada di kolom tersebut

test.dropna(subset=['tax liens'], inplace=True)
test.dropna(subset=['bankruptcies'], inplace=True)
```

```python
# Jumlah missing value pada kolom credit score dan kolom annual income sama, mari kita cek data tersebut
test[(test['credit score'].isna()) & (test['annual income'].isna()==False)]
```

Berdasarkan pengecekan, diketahui bahwa setiap missing value pada kolom credit score memiliki missing value juga pada kolom annual income.
Data tersebut dapat dihapus karena credit score dan annual income merupakan feature yang penting dalam menentukan loan status
